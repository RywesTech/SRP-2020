```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import time

starttime=time.time()

#fig.show()

#while True:
df = pd.read_csv('data/flight.csv')
fig = go.FigureWidget()
fig.add_trace(go.Scatter(x=df['ms'], y=df['0_pos'], name='Position', line=dict(color='firebrick', width=2)))
fig.add_trace(go.Scatter(x=df['ms'], y=df['0_output'], name = 'TVC Angle', line=dict(color='royalblue', width=2)))
fig.add_trace(go.Scatter(x=df['ms'], y=df['0_p'], name = 'P', line=dict(color='red', width=2)))
fig.add_trace(go.Scatter(x=df['ms'], y=df['0_i'], name = 'I', line=dict(color='green', width=2)))
fig.add_trace(go.Scatter(x=df['ms'], y=df['0_d'], name = 'D', line=dict(color='blue', width=2)))
fig.show()
#time.sleep(1.0 - ((time.time() - starttime) % 1.0))
```
